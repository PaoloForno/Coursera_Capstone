# Tecnical features

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('/content/drive/MyDrive/Project/dataset/train.csv')
df.head()
```

```
for col_name in df.columns: 
    print(col_name)
```

Equipment category

```
df_equipment = df[[col for col in df.columns if 'equipment_' in col]]
df_equipment = df_equipment.reindex(sorted(df_equipment.columns), axis=1)
df_equipment.head()
```

```
plt.figure(figsize=(12,9))
sns.heatmap(df_equipment.corr(), square=True, annot=True, cmap='Blues')
```

Only equipment_sum_alarms_prev14d

Correlation between SUM_ALARMS variables
